```python
from selenium import webdriver
import os
from bs4 import BeautifulSoup
import bs4
import pandas as pd
import time
import csv
import urllib.request as u
```

```python
###의료정책 크롤링
```

```python
#1_크롬_드라이버_가져오기
browser=webdriver.Chrome('chromedriver.exe')
```

```python
#2_언론사_페이지와_검색어_불러오기_의료정책
browser.get('http://www.ichannela.com/com/cmm/total_search.do?search='+'의료정책')
```

```python
#3_페이지_넘어가며_링크불러오기_의료정책
url=[]
p=0
browser.get('http://www.ichannela.com/com/cmm/total_search.do?search='+'의료정책')
while p < 21:
    browser.find_element_by_xpath("/html/body/div/div[2]/div/div/div[4]/a[1]/span").click()
    p += 1
```

```python
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
attr=soup.select('ul.list-search > li > a') 

for urls in attr:
    url.append(urls['href'].replace('\n','').replace('\t',''))
url=url[0:60]
```

```python
url
```

```python
#4_기사제목과_본문추출_의료정책
dict={}
for i in range(len(url)):
    browser.get("http://www.ichannela.com/"+url[i])
    time.sleep(5)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')

    #기사제목
    title=soup.find('strong',{'class':'subject_pg'}).text
    title=title.replace('\r','').replace('\n','').replace('\t','')

    #기사 본문 추출
    text=[]
    for content in soup.find('div',{'id':'articleSection'}):
        if type(content) is bs4.element.NavigableString and len(content)>20:
            #print(type(content))
            #print(content)
            text.append(content.replace('\xa0',' ').replace('\n','').replace('\t',''))
    #print(text)

    dict[title]=text
```

```python
#5_파일쓰기_의료정책
with open('CHANNELA_의료정책.csv','w',-1,encoding='utf-8-sig') as f:
    w=csv.writer(f)
    w.writerow(dict.keys())
    w.writerow(dict.values())
```

```python
#6_파일저장_의료정책
a=pd.read_csv('C:/Users/User/CHANNELA_의료정책.csv',encoding='utf-8-sig')
a=a.transpose() #현재 a는 2행 10열로 행에는 제목, 내용이 각 열에 들어있다. 이를 형변환 하여 10행 2열로 기사마다 [제목,본문]컬럼이다.
b=list
b=list
title=b(a.index)
c=list
text=a[0]
text=c(text)
source=pd.DataFrame({'Title':title,'Text':text})
source=source.iloc[:60]
source['Press']='CHANNELA'
source=source[['Press','Title','Text']]
source.to_csv('CHANNELA_의료정책.csv',encoding='utf-8-sig')
```

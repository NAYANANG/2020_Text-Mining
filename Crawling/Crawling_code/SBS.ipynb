```python
from selenium import webdriver
import os
from bs4 import BeautifulSoup
import bs4
import pandas as pd
import time
import csv
import urllib.request as ur
```

```python
#1_크롬_드라이버_가져오기
browser=webdriver.Chrome('chromedriver.exe')
```

```python
###의료정책 크롤링
```

```python
#2_언론사_페이지와_검색어_불러오기_의료정책
browser.get('https://news.sbs.co.kr/news/search/main.do?pageIdx=1&searchTermStartDate=&searchTermEndDate=&searchSection=01%7C02%7C03%7C07%7C08%7C09%7C14&searchCategory=&searchMode=&query={}&collection=nnews&searchOption=1&sort=RANK'.format('의료정책'))
```

```python
#3_페이지_넘어가며_링크불러오기_의료정책
url=[]
for p in range(1,7):
    browser.get('https://news.sbs.co.kr/news/search/main.do?pageIdx={}&searchTermStartDate=&searchTermEndDate=&searchSection=01%7C02%7C03%7C07%7C08%7C09%7C14&searchCategory=&searchMode=&query={}&collection=nnews&searchOption=1&sort=RANK'.format(p,'의료정책'))
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')
    attr=soup.select('div.psil_inner> a') 
    
    for urls in attr:
        url.append(urls['href'])
```

```python
url
```

```python
#4_기사제목과_본문추출_의료정책
dict={}
for i in range(len(url)+1):
    browser.get(url[i])
    time.sleep(5)
    html=browser.page_source
    soup=BeautifulSoup(html,'html.parser')

    #기사제목
    title=soup.find('h3',{'id':'vmNewsTitle'}).text
    title=title.replace('\r','').replace('\n','').replace('\t','')

    #기사 본문 추출
    text=[]
    for content in soup.find('div',{'class':'text_area'}):
        if type(content) is bs4.element.NavigableString and len(content)>20:
            #print(type(content))
            #print(content)
            text.append(content.replace('\xa0',' ').replace('\n',''))
    print(text)

    dict[title]=text
```

```python
#5_파일쓰기_의료정책
with open('SBS_의료정책.csv','w',-1,encoding='euc-kr') as f:
    w=csv.writer(f)
    w.writerow(dict.keys())
    w.writerow(dict.values())
```

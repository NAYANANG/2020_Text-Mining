```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
```

```python
def house_doc(doc):
    import pickle
    with open('./부동산정책_Doc/{}_부동산정책_Noun_doc.txt'.format(doc),'rb') as f:
            data=pickle.load(f)
    char=''
    for i in range(len(data)):
        char+=data[i]
    with open('./{}_부동산정책_total.txt'.format(doc),'wb') as f:
            pickle.dump(char,f)
```

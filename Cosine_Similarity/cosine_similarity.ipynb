```python
text=[]
text.append(KTV)
text.append(korean_eco)
text.append(channel_a)
text.append(han)
text.append(yonhap)
text.append(world)
text.append(munwha)
text.append(kukmin)
text.append(tv_chosun)
text.append(sbs_cnbc)
text.append(sbs)
text.append(news1)
text.append(mbn)
text.append(mbc)
```

```python
vect=TfidfVectorizer()
data=vect.fit_transform(text)
cosine_similarity_matrix=(data*data.T)
print(cosine_similarity_matrix.shape)
cosine_array=cosine_similarity_matrix.toarray()
```

```python
from konlpy.tag import Komoran

komoran=Komoran()
```

```python
import pandas as pd
import os
import numpy as np
KTV_rs=pd.read_csv('/home/u1013/schtm_osy_lsm_kjh/Crawling_Dataset/KTV국민방송_부동산정책.csv')
```

```python
KTV_rs=KTV_rs.drop('Unnamed: 0',axis=1)
```

```python
KTV_rs_text=np.array(KTV_rs.loc[:,'Text'].tolist())
```

```python
KTV_rs_noun_words=[]
for i in range(len(KTV_rs)):
    KTV_rs_noun_words.append(komoran.nouns(KTV_rs_text[i]))
```

```python
KTV_rs_noun_words
```

```python
stopwords=['앵커','기자','KTV','국민방송','케이블','방송','영상','위성방송','한국정책방송원']
for i in range(len(KTV_rs)):
    unique_Noun_words=set(KTV_rs_noun_words[i])
    for word in unique_Noun_words:
        if word in stopwords:
            while word in KTV_rs_noun_words[i]:KTV_rs_noun_words[i].remove(word)
```

```python
from collections import Counter
c=Counter(KTV_rs_noun_words[5])
print(c.most_common(10))
```

```python
from selenium import webdriver
import os
from bs4 import BeautifulSoup
import pandas as pd
import time
import csv
```

```python
 #크롬 드라이버 가져오기
browser=webdriver.Chrome('chromedriver.exe')
#세계일보 홈페이지 입장
browser.get('https://www.segye.com/search')
```

```python
#keyword 입력후 기사 본문 가져오기
browser.find_element_by_id('searchWord').send_keys('코로나방역')
browser.find_element_by_xpath('//*[@id="wps_layout1_box1"]/div/div[2]/img[1]').click()
time.sleep(10)
browser.find_element_by_xpath('//*[@id="articleTitleArea"]/div/a').click()
time.sleep(10)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
attr=soup.select('ul.listBox > li > a')
    
#기사 본문에서 url링크들 가져와 list로 만들기
url=[]
for urls in attr:
    url.append(urls['href'])
browser.find_element_by_xpath('//*[@id="pagingNav"]/a[2]').click()
time.sleep(2)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
attr=soup.select('ul.listBox > li > a') 
for urls in attr:
    url.append(urls['href'])
browser.find_element_by_xpath('//*[@id="pagingNav"]/a[4]').click()
time.sleep(2)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
attr=soup.select('ul.listBox > li > a') 
for urls in attr:
    url.append(urls['href'])
browser.find_element_by_xpath('//*[@id="pagingNav"]/a[5]').click()
time.sleep(2)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
attr=soup.select('ul.listBox > li > a') 
for urls in attr:
    url.append(urls['href'])
browser.find_element_by_xpath('//*[@id="pagingNav"]/a[6]').click()
time.sleep(2)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
attr=soup.select('ul.listBox > li > a') 
for urls in attr:
    url.append(urls['href'])
browser.find_element_by_xpath('//*[@id="pagingNav"]/a[7]').click()
time.sleep(2)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
attr=soup.select('ul.listBox > li > a') 
for urls in attr:
    url.append(urls['href'])
browser.find_element_by_xpath('//*[@id="pagingNav"]/a[8]').click()
time.sleep(2)
html=browser.page_source
soup=BeautifulSoup(html,'html.parser')
attr=soup.select('ul.listBox > li > a') 
for urls in attr:
    url.append(urls['href'])
```

```python
dict={}
for i in range(len(url)):
    browser.get(url[i])
    html=browser.page_source
    time.sleep(2)
    soup=BeautifulSoup(html,'html.parser')
    title=soup.find('h3',{'id':'title_sns'}).text
    source=soup.find('article',{'class':'viewBox'}).text
    source=source.replace('\n','').replace('\xa0','')
    dict[title]=source
#dictionary파일을 csv파일로 저장해준다
with open('COVID_세계일보.csv','w',-1,encoding='utf-8') as f:
    w=csv.writer(f)
    w.writerow(dict.keys())
    w.writerow(dict.values())
    
#csv파일들을 불러와 수정해준다.
world=pd.read_csv('./COVID_세계일보.csv',encoding='utf-8')
#행렬전환
world=world.transpose()
list_b=list
title=list_b(world.index)
list_c=list
text=world[0]
text=list_c(text)
source=pd.DataFrame({'Title':title,'Text':text})
#60개의 기사들만 추출한다.
source=source.iloc[:100]
#Press열에 언론사 이름을 넣어 붙여준다.
source['Press']='세계일보'
source=source[['Press','Title','Text']]
source.to_csv('COVID_세계일보.csv',encoding='utf-8-sig')
```

```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from math import log # IDF 계산을 위해
```

```python
with open('/home/u1009/schtm_osy_lsm_kjh/POS_Tag/SBS_코로나방역_Noun_doc.txt','rb') as f:
    Doc=pickle.load(f)
```

```python
Doc
```

```python
vocab = list(set(w for doc in Doc for w in doc.split()))
vocab.sort()
```

```python
#TF, IDF, 그리고 TF-IDF 값을 구하는 함수
N = len(Doc) # 총 문서의 수

def tf(t, d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in Doc:
        df += t in doc
    return log(N/(df + 1))

def tfidf(t, d):
    return tf(t,d)* idf(t)
```

```python
#DTM을 데이터프레임에 저장하여 출력
result = []
for i in range(N): # 각 문서에 대해서 아래 명령을 수행
    result.append([])
    d = Doc[i]
    for j in range(len(vocab)):
        t = vocab[j]        
        result[-1].append(tf(t, d))

tf_ = pd.DataFrame(result, columns = vocab)
tf_
```

```python
#각 단어에 대한 IDF 값
result = []
for j in range(len(vocab)):
    t = vocab[j]
    result.append(idf(t))

idf_ = pd.DataFrame(result, index = vocab, columns = ["IDF"])
idf_
```

```python
#TF-IDF 행렬
result = []
for i in range(N):
    result.append([])
    d = Doc[i]
    for j in range(len(vocab)):
        t = vocab[j]

        result[-1].append(tfidf(t,d))

tfidf_ = pd.DataFrame(result, columns = vocab)
tfidf_
```

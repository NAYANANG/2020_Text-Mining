```python
import pandas as pd
import numpy as np
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
 def TF_IDF(keyword):
    
    with open('/home/u1009/schtm_osy_lsm_kjh/POS_Tag/{0}_Noun_doc.txt'.format(keyword),'rb') as f:
        Doc=read=pickle.load(f,encoding='ecu-kr')
    tfidfv=TfidfVectorizer(100,0.95,0)
    tfidf_fit=tfidfv.fit_transform(Doc)
    tfidfv_dict=tfidfv.get_feature_names()
    tf_idf_array=tfidf_fit.toarray()
    tf_idf_df=pd.DataFrame(tf_idf_array,columns=tfidfv_dict)
    tf_idf_df.to_pickle('{}.csv'.format(keyword))
```

```python
#TF, IDF, 그리고 TF-IDF 값을 구하는 함수
N = len(Doc) # 총 문서의 수

def tf(t, d):
    return d.count(t)

def idf(t):
    df = 0
    for doc in Doc:
        df += t in doc
    return log(N/(df + 1))

def tfidf(t, d):
    return tf(t,d)* idf(t)
```

```python
import pickle

def n_gram(keyword):
    with open('/home/u1016/schtm_osy_lsm_kjh/POS_Tag/{0}_Noun_doc.txt'.format(keyword),'rb') as f:
        Doc=read=pickle.load(f)
    n_gram=[]
    for i in range(len(Doc)):
        n_sub=[]
        words=Doc[i].split()
        for j in range(len(words)-1):
            n_sub.append(words[j]+words[j+1])
        n_gram.append(n_sub)

    with open('{0}_n_gram.txt'.format(keyword),'wb') as f:
        pickle.dump(n_gram,f)     
    with open('{0}_n_gram.txt'.format(keyword),'rb') as f:
        read=pickle.load(f)
        print(read) 
```
